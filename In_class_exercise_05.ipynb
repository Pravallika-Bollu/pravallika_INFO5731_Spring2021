```python
# Please write your code here
import requests
import re
import nltk
nltk.download()
from nltk.corpus import stopwords
from textblob import TextBlob
from nltk.stem import PorterStemmer
from textblob import Word
from bs4 import BeautifulSoup
import pandas as pd
nltk.download('wordnet')

st = PorterStemmer()

main = []
sub =[]
for number in range(50):
  link = "https://www.flipkart.com/samsung-80-cm-32-inch-hd-ready-led-smart-tv-2020/p/itma5db0294e13fe?pid=TVSFQXFKMHUDYCJ3&lid=LSTTVSFQXFKMHUDYCJ3IOJXHJ&marketplace=FLIPKART&srno=b_1_1&otracker=browse&fm=organic&iid=8f1547ff-b24b-4415-b1b2-9cc940fe4743.TVSFQXFKMHUDYCJ3.SEARCH&ppt=clp&ppn=mobile-phones-store&ssid=j3wcntaq340000001615003131840" + str(number) # Generating link dynamically
  page=requests.get(link)
  soup = BeautifulSoup(page.text, 'html.parser')
  main_reviews = soup.find_all(class_='_2-N8zT') 
  text_reviews = soup.find_all(class_='t-ZTKy') 
  for ele, sub_ele in zip(main_reviews, text_reviews) :
      main.append(ele.text)
      sub.append(sub_ele.text)
df = pd.DataFrame(list(zip(main, sub)), columns =['Inshort', 'Full Review'])
print("Length of data frame is {0}".format(len(df)))
#converting to lower case
df['Lower case'] = df['Full Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
# Removing Punctuation
df['Removal of Punctuation'] = df['Lower case'].str.replace('[^\w\s]','')
#Removing Numerics
df['Removal of Numerics'] = df['Removal of Punctuation'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
#Removing Special Characters
df['Removal of special Characters'] = df['Removal of Numerics'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
# Removal of stop words
stop = stopwords.words('english')
df['Removal of stop words'] = df['Removal of special Characters'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
#spelling correction
df['spelling correction'] = df['Removal of stop words'].apply(lambda x: str(TextBlob(x).correct()))
#stemming
df['stemming'] = df['Removal of stop words'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
#Lemmatization
df['Lemmatization'] = df['Removal of stop words'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df
```

```python
# PARTS OF SPEECH TAGGING AND FEATURES

from nltk.tokenize import word_tokenize
pos = []
for sentence in df['Lemmatization']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos
```

```python
Adjective = []
Adverb = []
CordinatingConjunction = []
SubordinatingConjuction = []
Interjection = []
Noun = []
Verb = []
PersonalPronoun = []
predeterminer = []
Determiner = []
for value in pos:
  AdjectiveCount = 0
  AdverbCount = 0
  CordinatingConjunctionCount = 0
  SubordinatingConjuctionCount = 0
  InterjectionCount = 0
  NounCount = 0
  VerbCount = 0
  PersonalPronounCount = 0
  predeterminerCount = 0
  DeterminerCount = 0
  for word,tag in value:
    if tag == 'JJ':
      AdjectiveCount = AdjectiveCount + 1
    elif tag == 'RB':
      AdverbCount = AdverbCount + 1
    elif tag == 'CC':
      CordinatingConjunctionCount = CordinatingConjunctionCount + 1
    elif tag == 'UH':
      InterjectionCount = InterjectionCount + 1
    elif tag == 'NN':
      NounCount = NounCount + 1
    elif tag == 'VR':
      VerbCount = VerbCount + 1
    elif tag == 'PRP':
      PersonalPronounCount = PersonalPronounCount + 1
    elif tag == 'PDT':
      predeterminerCount = predeterminerCount + 1
    elif tag == 'DT':
      DeterminerCount = DeterminerCount + 1
    elif tag == 'IN':  
      SubordinatingConjuctionCount = SubordinatingConjuctionCount + 1
  Adjective.append(AdjectiveCount)
  Adverb.append(AdverbCount)
  CordinatingConjunction.append(CordinatingConjunctionCount)
  Interjection.append(InterjectionCount)
  Noun.append(NounCount)
  Verb.append(VerbCount)
  PersonalPronoun.append(PersonalPronounCount)
  predeterminer.append(predeterminerCount)
  Determiner.append(DeterminerCount)
  SubordinatingConjuction.append(SubordinatingConjuctionCount)
```

```python
df['Number of Adjectives'] = Adjective
df['Number of Adverbs'] = Adverb
df['Number of Cordinating Conjunctions'] = CordinatingConjunction
df['Number of Interjections'] = Interjection
df['Number of Nouns'] = Noun
df['Number of Verbs'] = Verb
df['Number of Personal Pronouns'] = PersonalPronoun
df['Number of Predeterminers'] = predeterminer
df['Number of Determiners'] = Determiner
df['Number of Subordinating Conjuctions'] = SubordinatingConjuction
df
```
